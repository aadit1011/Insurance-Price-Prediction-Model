Region VS Children

```python
# Encode the 'region' column
df['region_encoded'] = df['region'].astype('category').cat.codes

# Create a heatmap for children and region_encoded
heatmap_data = df[['children', 'region_encoded']].corr()  # Compute correlation matrix

sns.heatmap(heatmap_data, annot=True, cmap='coolwarm')
plt.title('Heatmap: Children vs Region (Encoded)')
plt.savefig('Heatmap: Children vs Region (Encoded)')
plt.show()
```

Smoker VS Sex 

```python
# Encode the 'smoker' column
df['smoker_encoded'] = df['smoker'].astype('category').cat.codes
# Encode the 'sex' column
df['sex_encoded'] = df['sex'].astype('category').cat.codes

# Create a heatmap for children and region_encoded
heatmap_data = df[['sex_encoded', 'smoker_encoded']].corr()  # Compute correlation matrix

sns.heatmap(heatmap_data, annot=True, cmap='coolwarm')
plt.title('Heatmap: Sex vs Smoker (Encoded)')
plt.savefig('Heatmap: Sex vs Smoker (Encoded)')
plt.show()
```
